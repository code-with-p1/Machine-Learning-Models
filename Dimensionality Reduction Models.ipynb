### Setup and Common Code

```python
# Import all necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings('ignore')
# Import Datasets
from sklearn.datasets import load_digits, load_iris, fetch_olivetti_faces

# Function to visualize 2D projections
def plot_2d_projection(X_transformed, y, title):
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y, 
                         cmap='viridis', s=50, alpha=0.8)
    plt.colorbar(scatter)
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.show()

# Function to visualize image data reconstruction (for PCA/NMF)
def plot_reconstruction(original, reconstructed, title, n_images=5):
    fig, axes = plt.subplots(2, n_images, figsize=(2*n_images, 4))
    for i in range(n_images):
        # Original
        axes[0, i].imshow(original[i].reshape(64, 64), cmap='gray')
        axes[0, i].set_title(f"Original {i+1}")
        axes[0, i].axis('off')
        
        # Reconstructed
        axes[1, i].imshow(reconstructed[i].reshape(64, 64), cmap='gray')
        axes[1, i].set_title(f"Reconstructed {i+1}")
        axes[1, i].axis('off')
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

### Dataset Preparation

```python
# Option A: load_digits (64-dimensional -> 8x8 pixels)
digits = load_digits()
X_digits, y_digits = digits.data, digits.target
X_digits_scaled = StandardScaler().fit_transform(X_digits)
dataset_name = "Digits"

# Option B: load_iris (4-dimensional)
iris = load_iris()
X_iris, y_iris = iris.data, iris.target
X_iris_scaled = StandardScaler().fit_transform(X_iris)
dataset_name = "Iris"

# Option C: fetch_olivetti_faces (4096-dimensional -> 64x64 pixels)
faces = fetch_olivetti_faces()
X_faces, y_faces = faces.data, faces.target
# Note: Faces are already normalized between 0-1, but we can still scale
X_faces_scaled = StandardScaler().fit_transform(X_faces)
dataset_name = "Faces"
```

# Dimensionality Reduction Models

### 1. Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

# Choose a dataset (X) from the options above
X = X_iris_scaled
y = y_iris

# Create and fit the model
print("Principal Component Analysis (PCA):")
model = PCA(n_components=2, random_state=42)  # Reduce to 2 dimensions
X_transformed = model.fit_transform(X)

print(f"Explained variance ratio: {model.explained_variance_ratio_}")
print(f"Total explained variance: {sum(model.explained_variance_ratio_):.3f}")

# Visualize the 2D projection
plot_2d_projection(X_transformed, y, 'PCA Projection (2D)')
```

### 2. t-Distributed Stochastic Neighbor Embedding (t-SNE)

```python
from sklearn.manifold import TSNE

# Choose a dataset (X) from the options above
X = X_digits_scaled
y = y_digits

# Create and fit the model
print("t-SNE:")
# Note: t-SNE is stochastic, so results vary each time. perplexity is a key parameter.
model = TSNE(n_components=2, random_state=42, perplexity=30)
X_transformed = model.fit_transform(X)

# Visualize the 2D projection
plot_2d_projection(X_transformed, y, 't-SNE Projection (2D)')
```

### 3. Non-negative Matrix Factorization (NMF)

```python
from sklearn.decomposition import NMF

# Choose a dataset (X) from the options above - best for non-negative data like faces
X = X_faces
y = y_faces

# Create and fit the model
print("Non-negative Matrix Factorization (NMF):")
model = NMF(n_components=2, random_state=42, init='nndsvd')
X_transformed = model.fit_transform(X)

# Reconstruct the images to see the quality
X_reconstructed = model.inverse_transform(X_transformed)

# Visualize the 2D projection
plot_2d_projection(X_transformed, y, 'NMF Projection (2D)')

# For faces, also show reconstruction quality
plot_reconstruction(X[:5], X_reconstructed[:5], 'NMF Face Reconstruction')
```

### 4. Truncated SVD

```python
from sklearn.decomposition import TruncatedSVD

# Choose a dataset (X) from the options above
X = X_digits_scaled
y = y_digits

# Create and fit the model
print("Truncated SVD:")
model = TruncatedSVD(n_components=2, random_state=42)
X_transformed = model.fit_transform(X)

print(f"Explained variance ratio: {model.explained_variance_ratio_}")
print(f"Total explained variance: {sum(model.explained_variance_ratio_):.3f}")

# Visualize the 2D projection
plot_2d_projection(X_transformed, y, 'Truncated SVD Projection (2D)')
```

## Complete Example: Comparing Techniques on Digits Dataset

```python
# --- COMPLETE EXAMPLE: Comparing Dimensionality Reduction on Digits ---
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, NMF, TruncatedSVD
from sklearn.manifold import TSNE

# Load and prepare data
digits = load_digits()
X, y = digits.data, digits.target
X_scaled = StandardScaler().fit_transform(X)

# Create models
models = {
    'PCA': PCA(n_components=2, random_state=42),
    't-SNE': TSNE(n_components=2, random_state=42, perplexity=30),
    'NMF': NMF(n_components=2, random_state=42, init='nndsvd'),
    'Truncated SVD': TruncatedSVD(n_components=2, random_state=42)
}

# Plot results
plt.figure(figsize=(15, 12))

for i, (name, model) in enumerate(models.items(), 1):
    if name == 'NMF':
        # NMF requires non-negative data
        X_transformed = model.fit_transform(X - X.min())
    else:
        X_transformed = model.fit_transform(X_scaled)
    
    plt.subplot(2, 2, i)
    scatter = plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y, 
                         cmap='viridis', s=30, alpha=0.8)
    plt.colorbar(scatter)
    plt.title(f'{name} Projection')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')

plt.tight_layout()
plt.show()

# Print explained variance for linear methods
print("Explained Variance Ratios:")
pca = PCA(n_components=2).fit(X_scaled)
svd = TruncatedSVD(n_components=2).fit(X_scaled)
print(f"PCA: {sum(pca.explained_variance_ratio_):.3f}")
print(f"Truncated SVD: {sum(svd.explained_variance_ratio_):.3f}")
```

### Special Example: PCA for Image Reconstruction (Faces)

```python
# --- PCA for Face Reconstruction ---
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces

# Load faces data
faces = fetch_olivetti_faces()
X, y = faces.data, faces.target

# Apply PCA with more components to see reconstruction quality
n_components = 350  # Reconstruct with 100 principal components
pca = PCA(n_components=n_components, random_state=42)
X_pca = pca.fit_transform(X)
X_reconstructed = pca.inverse_transform(X_pca)

print(f"Explained variance with {n_components} components: {sum(pca.explained_variance_ratio_):.3f}")

# Visualize original vs reconstructed faces
plot_reconstruction(X, X_reconstructed, f'PCA Face Reconstruction ({n_components} components)', n_images=5)
```
